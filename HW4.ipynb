```python
import talib
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
data = yf.download('2454.TW',start = '2024-01-01')
data['Close'].values
arr = []
for num in data['Close'].values:
    for i in num:
        arr.append(i)
arrnp = np.array(arr)
rsi6   = talib.RSI(arrnp, timeperiod=6)  #短天期 RSI
rsi14  = talib.RSI(arrnp, timeperiod=14) #長天期 RSI

flag = rsi6[0] - rsi14[0]
for i in range(len(rsi6)):
    temp = rsi6[i] - rsi14[i]
    if flag * temp < 0:
        print(rsi6[i-1],rsi14[i-1])
        print(rsi6[i],rsi14[i])
        if flag > temp:
            print("RSI出現死亡交叉 給予建議Sell",end='\n' )
        else:
            print("RSI出現黃金交叉 給予建議Buy",end='\n')
    flag = temp
```
