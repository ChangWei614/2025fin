```python
# 導入相關函式庫
import os
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

os.environ['TEJAPI_BASE']='https://api.tej.com.tw'
os.environ['TEJAPI_KEY'] ='TEJ API Key'

# 設定回測起訖時間與股票代碼
start = '2013-01-01'
end = '2024-12-31'
os.environ['mdate'] = start + ' ' + end

# 設定投資期間全部的股票池
os.environ['ticker'] = ' '.join(['2611', '1435', '3054', '1339', '2488', '1418', '2443', '9935', '3652', '1524', '3048', '1538', '1453', '3060', '1452', '2472', '2453', '1616', '6142', '2017', '1446', '2302', '1441', '1467', '2365', '2484', '6133', '1417', '8011', '1618', '1454', '2375', '1432', '3591', '1473', '2530', '2359', '8249', '1472', '1906', '2493', '3550', '1540', '3383', '1443', '6226', '2431', '1517', '6141', '6225', '9801', '2468', '1617', '1438', '3002', '1225', '2364', '1514', '1603', '2613', '6164', '3029', '2390', '1456', '2413', '3536', '1471', '5484', '2537', '3018', '1466', '3701', '2906', '1215', '1713', '5515', '2904', '2440', '2417', '2427', '4414', '1463', '2342', '3027', '1464', '2491', '1445', '1515', '2460', '1465', '1528', '1516', '3579', '2032', '1457', '5519', '1526', '9955', '8103', '3305', '3024', '9912', '2494', '3052', '2543', '1325', '2516', '9929', '2419', '2024', '3669', '3056', '3557', '1529', '3315', '6117', '6152', '1413', '1423', '3055', '6136', '2461', '3051', '1469', '2361', '3311', '5608', '1615', '3038', '3312', '8201', '1474', '4930', '9906', '2442', '3050', '2438', '3419', '2305', '9944', '1809', '1410', '3032', '2399', '1207', '2477', '3432', '2546', '2429', '1235', '2008', '1219', '3046', '3021', '2025', '1449', '3607', '9902', '3058', '3011', '2033', '1439', '8438', '8072', '2528', '1316', '2022', '6405', '3622', '1414', '6112', '8021','IR0001'])

# 輸入到tquant資料包中
!zipline ingest -b tquant

# 導入zipline回測套件
from zipline.api import *
from zipline.finance import commission, slippage

def initialize(context):
    context.yearly_stocks = {
    2013: ['1443', '1445', '1524', '1906', '2361', '2443', '2484', '3032', '3046', '3312', '3432', '3550', '8249', '9944'],
    2014: ['1413', '1449', '1452', '1469', '1906', '2024', '2025', '2302', '2413', '2443', '2468', '2516', '2906', '3011', '3024', '3038', '3046', '3315', '6117', '6133', '6141', '8249'],
    2015: ['1445', '1454', '1474', '1515', '2017', '2375', '3050', '6133'],
    2016: ['1235', '1439', '1456', '1474', '1514', '1538', '1809', '1906', '2008', '2022', '2024', '2025', '2032', '2305', '2375', '2399', '2417', '3050', '3058', '3669', '5484', '6152', '6226', '9955'],
    2017: ['1235', '1423', '1435', '1453', '2399', '2427', '2453', '2460', '3055', '3383', '3579', '3622', '8201'],
    2018: ['1417', '1423', '1435', '1441', '1443', '1454', '1471', '1538', '2033', '2302', '3011', '3052', '3432', '5484', '5519', '6142', '8072', '9912'],
    2019: ['1417', '1456', '2431', '2453', '3002', '3024', '3052', '3305', '3311', '3383', '5484', '6152'],
    2020: ['1414', '1423', '1432', '1445', '1471', '1474', '1517', '1616', '1618', '2033', '2305', '2365', '2611', '2613', '3058', '3591', '4930', '5608', '6164', '6405', '8011', '8072', '8103'],
    2021: ['1445', '1452', '1454', '1456', '2611', '3060', '6405'],
    2022: ['1339', '1417', '1418', '1423', '1449', '1465', '1473', '1540', '1618', '2442', '2488', '2543', '3052', '3432', '3652', '4930'],
    2023: ['1325', '1339', '1418', '1423', '2022', '2025', '2365', '2443', '3591', '3607', '5515', '8103'],
    2024: ['1423', '1439', '1528', '1809', '2431', '2477', '2537', '3550', '6405', '8021', '8438', '9902']
    }

    # 設定各年度首個交易日
    context.first_trading_days = {
        2013: pd.Timestamp('2013-01-02').date(),
        2014: pd.Timestamp('2014-01-02').date(),
        2015: pd.Timestamp('2015-01-05').date(),
        2016: pd.Timestamp('2016-01-04').date(),
        2017: pd.Timestamp('2017-01-03').date(),
        2018: pd.Timestamp('2018-01-02').date(),
        2019: pd.Timestamp('2019-01-02').date(),
        2020: pd.Timestamp('2020-01-02').date(),
        2021: pd.Timestamp('2021-01-04').date(),
        2022: pd.Timestamp('2022-01-03').date(),
        2023: pd.Timestamp('2023-01-03').date(),
        2024: pd.Timestamp('2024-01-02').date(),
    }

    # 設定各年度最後一個交易日
    context.last_trading_days = {
        2013: pd.Timestamp('2013-12-31').date(),
        2014: pd.Timestamp('2014-12-31').date(),
        2015: pd.Timestamp('2015-12-31').date(),
        2016: pd.Timestamp('2016-12-30').date(),
        2017: pd.Timestamp('2017-12-29').date(),
        2018: pd.Timestamp('2018-12-28').date(),
        2019: pd.Timestamp('2019-12-31').date(),
        2020: pd.Timestamp('2020-12-31').date(),
        2021: pd.Timestamp('2021-12-30').date(),
        2022: pd.Timestamp('2022-12-30').date(),
        2023: pd.Timestamp('2023-12-29').date(),
        2024: pd.Timestamp('2024-12-21').date(),
    }

    # Buy/Sell flags設定
    years = range(2013,2025)
    context.buy_flags  = {year: False for year in years}
    context.sell_flags = {year: False for year in years}

    # 設定滑價
    set_slippage(slippage.VolumeShareSlippage(volume_limit=0.025, price_impact=0.1))
    # 設定交易手續費
    set_commission(commission.Custom_TW_Commission(min_trade_cost=20, discount=1.0, tax=0.003))
    # 設定大盤指數
    set_benchmark(symbol('IR0001'))

def handle_data(context, data):
    # 忽略時區&時間資訊，僅得到YYYY/MM/DD
    current_date = get_datetime().date()
    # 得到YYYY/MM/DD後，取得YYYY
    current_year = current_date.year

    # 年初買入
    if current_date == context.first_trading_days.get(current_year) and not context.buy_flags[current_year]:
        tickers = context.yearly_stocks[current_year]
        tradable = [tk for tk in tickers if data.can_trade(symbol(tk))]

        if tradable:
            # 採均等權重配置
            equal_weight = 1.0 / len(tradable)

            # 買進邏輯：每年初重新下單，採均等權重配置
            for ticker in tradable:
                order_target_percent(symbol(ticker), equal_weight)
            
            context.buy_flags[current_year] = True
            print(f"{current_year} 年初買入完成，共 {len(tradable)} 檔，均等權重配置完成")

    # 年末賣出（全部清倉）
    elif current_date == context.last_trading_days.get(current_year) and not context.sell_flags[current_year]:
        for position in context.portfolio.positions:
            # 賣出邏輯：檢視年底持有股票池和持股總數，依據該持股總數賣出，避免超賣情形發生
            amount = context.portfolio.positions[position].amount
            if amount > 0:
                order(position,-amount)

        context.sell_flags[current_year] = True
        print(f"{current_year} 年末全部賣出")

def analyze(context, results):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    results['benchmark_cum'] = results.benchmark_return.add(1).cumprod() * capital_base
    results[['portfolio_value', 'benchmark_cum']].plot(ax=ax1)
    ax1.set_ylabel('Portfolio value (TWD)')
    plt.legend(['Portfolio', 'Benchmark'], loc='upper left')
    plt.gcf().set_size_inches(18, 8)
    plt.grid()
    plt.show()

from zipline import run_algorithm

capital_base = 1e6
start_date = pd.Timestamp(start, tz='utc')
end_date = pd.Timestamp(end, tz='utc')

results = run_algorithm(
    start=start_date,
    end=end_date,
    initialize=initialize,
    analyze=analyze,
    handle_data=handle_data,
    bundle='tquant',
    capital_base=capital_base)
results
```
